```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

def generate_realistic_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_users = random.randint(800, 1000)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.1
        elif 6 <= month <= 8:
            seasonal_factor = 1.2
        elif 9 <= month <= 11:
            seasonal_factor = 1.05
        else:
            seasonal_factor = 0.9
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-50, 50)
        
        size = int(base_users * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df


def generate_all_users(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_users = random.randint(10000, 12000)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.1
        elif 6 <= month <= 8:
            seasonal_factor = 1.2
        elif 9 <= month <= 11:
            seasonal_factor = 1.05
        else:
            seasonal_factor = 0.9
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-50, 50)
        
        size = int(base_users * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_churning_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_churn = random.randint(0, 30)
        
        if 3 <= month <= 5:
            seasonal_factor = 0.9
        elif 6 <= month <= 8:
            seasonal_factor = 0.8
        elif 9 <= month <= 11:
            seasonal_factor = 0.95
        else:
            seasonal_factor = 1.1
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-10, 10)
        
        size = int(base_churn * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_dormant_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_dormant = random.randint(200, 300)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.05
        elif 6 <= month <= 8:
            seasonal_factor = 1.08
        elif 9 <= month <= 11:
            seasonal_factor = 1.03
        else:
            seasonal_factor = 0.95
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-10, 10)
        
        size = int(base_dormant * seasonal_factor * (1 - trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_upsell_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_upsell = random.randint(10, 60)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.05
        elif 6 <= month <= 8:
            seasonal_factor = 1.08
        elif 9 <= month <= 11:
            seasonal_factor = 1.03
        else:
            seasonal_factor = 0.95
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-5, 5)
        
        size = int(base_upsell * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_new_users_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_new_users = random.randint(100, 200)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.2
        elif 6 <= month <= 8:
            seasonal_factor = 1.3
        elif 9 <= month <= 11:
            seasonal_factor = 1.1
        else:
            seasonal_factor = 0.9
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-10, 10)
        
        size = int(base_new_users * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_renewals_upcoming_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_upcoming_renewals = random.randint(30, 70)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.05
        elif 6 <= month <= 8:
            seasonal_factor = 1.1
        elif 9 <= month <= 11:
            seasonal_factor = 1.02
        else:
            seasonal_factor = 0.95
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-5, 5)
        
        size = int(base_upcoming_renewals * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_renewed_users_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_renewed_users = random.randint(20, 40)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.1
        elif 6 <= month <= 8:
            seasonal_factor = 1.2
        elif 9 <= month <= 11:
            seasonal_factor = 1.05
        else:
            seasonal_factor = 0.9
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-5, 5)
        
        size = int(base_renewed_users * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df


def generate_push_consent_enabled_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_push_consent = random.randint(5000, 9500)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.1
        elif 6 <= month <= 8:
            seasonal_factor = 1.2
        elif 9 <= month <= 11:
            seasonal_factor = 1.05
        else:
            seasonal_factor = 0.9
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-10, 10)
        
        size = int(base_push_consent * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_email_consent_enabled_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        month = date.month
        base_email_consent = random.randint(5000, 9500)
        
        if 3 <= month <= 5:
            seasonal_factor = 1.1
        elif 6 <= month <= 8:
            seasonal_factor = 1.15
        elif 9 <= month <= 11:
            seasonal_factor = 1.05
        else:
            seasonal_factor = 0.9
        
        trend = (date - pd.Timestamp("2022-11-01")).days / 180
        noise = random.uniform(-10, 10)
        
        size = int(base_email_consent * seasonal_factor * (1 + trend) + noise)
        size = max(size, 0)
        
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_global_control_group_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        size = int(0.1 * merged_data.loc[merged_data["date"] == date, "all_app_users"].values[0])
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_global_treatment_group_time_series(start_date, end_date):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        size = int(0.9 * merged_data.loc[merged_data["date"] == date, "all_app_users"].values[0])
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_feature_treatment_time_series(start_date, end_date, adoption_rate):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        size = int(0.9 * merged_data.loc[merged_data["date"] == date, "global_treatment_group"].values[0] * adoption_rate)
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

def generate_feature_control_time_series(start_date, end_date, adoption_rate):
    date_range = pd.date_range(start_date, end_date, freq='D')
    data = []

    for date in date_range:
        size = int(0.1 * merged_data.loc[merged_data["date"] == date, "global_control_group"].values[0] * adoption_rate)
        data.append([date, size])

    df = pd.DataFrame(data, columns=["date", "size"])
    return df

# Generate the time series data for different cohorts
start_date = "2022-11-01"
end_date = "2023-04-30"

# Set realistic adoption rates for each feature (between 0 and 1)
plant_added_adoption_rate = 0.05  # Example adoption rate
initial_purchase_adoption_rate = 0.15  # Example adoption rate
plant_article_viewed_adoption_rate = 0.25  # Example adoption rate
action_completed_adoption_rate = 0.10  # Example adoption rate



premium_users_data = generate_realistic_time_series(start_date, end_date)
all_users_data = generate_all_users(start_date, end_date)
churning_users_data = generate_churning_time_series(start_date, end_date)
dormant_users_data = generate_dormant_time_series(start_date, end_date)
upsell_users_data = generate_upsell_time_series(start_date, end_date)
new_users_data = generate_new_users_time_series(start_date, end_date)
upcoming_renewals_data = generate_renewals_upcoming_time_series(start_date, end_date)
renewed_users_data = generate_renewed_users_time_series(start_date, end_date)
push_consent_enabled_data = generate_push_consent_enabled_time_series(start_date, end_date)
email_consent_enabled_data = generate_email_consent_enabled_time_series(start_date, end_date)





# Merge the different cohort dataframes based on the date
merged_data = pd.merge(premium_users_data, churning_users_data, on="date", how="outer")
merged_data = pd.merge(merged_data, dormant_users_data, on="date", how="outer")
merged_data = pd.merge(merged_data, upsell_users_data, on="date", how="outer")
merged_data = pd.merge(merged_data, new_users_data, on="date", how="outer")
merged_data = pd.merge(merged_data, upcoming_renewals_data, on="date", how="outer")
merged_data = pd.merge(merged_data, renewed_users_data, on="date", how="outer")
merged_data = pd.merge(merged_data, all_users_data, on="date", how="outer")
merged_data = pd.merge(merged_data, push_consent_enabled_data, on="date", how="outer")
merged_data = pd.merge(merged_data, email_consent_enabled_data, on="date", how="outer")



# Rename columns for better clarity
merged_data.columns = [
    "date", 
    "premium_users", 
    "churning_users", 
    "dormant_users", 
    "upsell_users", 
    "new_users", 
    "upcoming_renewals", 
    "renewed_users", 
    "all_app_users", 
    "push_consent_enabled", 
    "email_consent_enabled"
]


global_control_group_data = generate_global_control_group_time_series(start_date, end_date)
global_treatment_group_data = generate_global_treatment_group_time_series(start_date, end_date)

merged_data = pd.merge(merged_data, global_control_group_data, on="date", how="outer")
merged_data = pd.merge(merged_data, global_treatment_group_data, on="date", how="outer")

merged_data.columns = [
    "date", 
    "premium_users", 
    "churning_users", 
    "dormant_users", 
    "upsell_users", 
    "new_users", 
    "upcoming_renewals", 
    "renewed_users", 
    "all_app_users", 
    "push_consent_enabled", 
    "email_consent_enabled", 
    "global_control_group", 
    "global_treatment_group"
]


# Generate the time series data for different app features with adoption rates
plant_added_control_data = generate_feature_control_time_series(start_date, end_date, plant_added_adoption_rate)
plant_added_treatment_data = generate_feature_treatment_time_series(start_date, end_date, plant_added_adoption_rate)

initial_purchase_control_data = generate_feature_control_time_series(start_date, end_date, initial_purchase_adoption_rate)
initial_purchase_treatment_data = generate_feature_treatment_time_series(start_date, end_date, initial_purchase_adoption_rate)

plant_article_viewed_control_data = generate_feature_control_time_series(start_date, end_date, plant_article_viewed_adoption_rate)
plant_article_viewed_treatment_data = generate_feature_treatment_time_series(start_date, end_date, plant_article_viewed_adoption_rate)

action_completed_control_data = generate_feature_control_time_series(start_date, end_date, action_completed_adoption_rate)
action_completed_treatment_data = generate_feature_treatment_time_series(start_date, end_date, action_completed_adoption_rate)

merged_data = pd.merge(merged_data, plant_added_control_data, on="date", how="outer")
merged_data = pd.merge(merged_data, plant_added_treatment_data, on="date", how="outer")
merged_data = pd.merge(merged_data, initial_purchase_control_data, on="date", how="outer")
merged_data = pd.merge(merged_data, initial_purchase_treatment_data, on="date", how="outer")
merged_data = pd.merge(merged_data, plant_article_viewed_control_data, on="date", how="outer")
merged_data = pd.merge(merged_data, plant_article_viewed_treatment_data, on="date", how="outer")
merged_data = pd.merge(merged_data, action_completed_control_data, on="date", how="outer")
merged_data = pd.merge(merged_data, action_completed_treatment_data, on="date", how="outer")


# Rename columns for better clarity
merged_data.columns = [
    "date", 
    "premium_users", 
    "churning_users", 
    "dormant_users", 
    "upsell_users", 
    "new_users", 
    "upcoming_renewals", 
    "renewed_users", 
    "all_app_users", 
    "push_consent_enabled", 
    "email_consent_enabled", 
    "global_control_group", 
    "global_treatment_group", 
    "plant_added_control", 
    "plant_added_treatment", 
    "initial_purchase_control", 
    "initial_purchase_treatment", 
    "plant_article_viewed_control", 
    "plant_article_viewed_treatment", 
    "action_completed_control", 
    "action_completed_treatment"
]



# Fill NaN values with 0 for missing data
merged_data = merged_data.fillna(0)

# Calculate "both push and email consent" segment by taking the minimum of the two segments
merged_data["both_push_email_consent"] = merged_data[["push_consent_enabled", "email_consent_enabled"]].min(axis=1)

# Calculate "no consent" segment by subtracting the sum of other consent segments from "all app users"
merged_data["no_consent"] = merged_data["all_app_users"] - merged_data[["push_consent_enabled", "email_consent_enabled", "both_push_email_consent"]].max(axis=1)


# Print the merged dataset
merged_data
```
